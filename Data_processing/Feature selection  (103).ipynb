## Chi square test

```python
df_new1 = df_tbi1.drop('record_id', axis = 1)
df_new = df_new1.drop('d_kl_tl', axis = 1)
df_new.head()
```

```python
import numpy as np
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

def chi_square_feature_selection(X, y, k=17):
    # Select the k best features using chi-square
    selector = SelectKBest(score_func=chi2, k=k)
    X_new = selector.fit_transform(X, y)
    # Get the chi-square scores
    chi_scores = selector.scores_
    # Get the indices of the selected features
    selected_indices = selector.get_support(indices=True)
    return selected_indices, chi_scores

# Example usage:
# Assuming df_new is your feature DataFrame and df_tbi1['d_kl_tl'] is your target variable.

# Load dataset as an example
X = df_new.values  # Assuming df_new is a DataFrame
y = df_tbi1['d_kl_tl'].values  # Assuming df_tbi1 is a DataFrame

# Apply chi-square feature selection
selected_indices, chi_scores = chi_square_feature_selection(X, y)

# Print the selected feature indices and their chi-square scores
print("Selected feature indices:", selected_indices)
print("Chi-square scores of selected features:")
for idx, score in zip(selected_indices, chi_scores[selected_indices]):
    print(f"Feature {idx}: {score}")
```

```python
import shap
from sklearn.ensemble import GradientBoostingClassifier
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split

# Assuming X is your feature DataFrame and y is your target variable (binary classification)
X = df_new.values  # Replace df_new with your actual feature DataFrame
y = df_tbi1['d_kl_tl'].values  # Replace df_tbi1['d_kl_tl'] with your actual target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# 1. GBM Feature Importance
def gbm_feature_importance(X, y):
    # Initialize Gradient Boosting Classifier
    gbm = GradientBoostingClassifier()
    
    # Fit the model
    gbm.fit(X_train, y_train)
    
    # Get feature importances
    feature_importances = gbm.feature_importances_
    
    # Sort feature importances in descending order
    indices = np.argsort(feature_importances)[::-1]
    
    # Print top 20 feature importances
    print("Top 20 Feature Importances:")
    for f in range(20):
        print("%d. Feature %d (%f)" % (f + 1, indices[f], feature_importances[indices[f]]))
    
    # Plot feature importances
    plt.figure(figsize=(10, 6))
    plt.title("GBM Feature Importances")
    plt.bar(range(20), feature_importances[indices][:20], color="skyblue", align="center")
    plt.xticks(range(20), indices[:20])
    plt.xlabel("Feature Index")
    plt.ylabel("Feature Importance")
    plt.show()

# 2. SHAP Analysis
def shap_analysis(X, y, feature_names=None):
    # Initialize Gradient Boosting Classifier
    gbm = GradientBoostingClassifier()
    
    # Fit the model
    gbm.fit(X_train, y_train)
    
    # Initialize a surrogate model for SHAP explanation
    surrogate_model = GradientBoostingRegressor()
    surrogate_model.fit(X_train, gbm.predict_proba(X_train)[:, 1])  # Predict probabilities for the positive class
    
    # Create a SHAP explainer object
    explainer = shap.TreeExplainer(surrogate_model)
    
    # Compute SHAP values for all samples in the training set
    shap_values = explainer.shap_values(X_train)
    
    # Get feature names if not provided
    if feature_names is None:
        feature_names = [f"Feature {i}" for i in range(X.shape[1])]
    
    # Summary plot of SHAP values
    shap.summary_plot(shap_values, X_train, plot_type="bar", feature_names=feature_names, show=False)
    
    # Add feature names to the plot
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

 


 # Assuming shap_values are already computed
# # Get feature names from the DataFrame
# feature_names = df_new.columns.tolist()

# #shap_analysis(X_train, y_train, feature_names=df_new.columns.tolist())            # Uncomment this line to perform SHAP analysis
# # Plot SHAP summary with feature names
# # Print list of feature names
# print("Feature Names:", feature_names)

# # Plot SHAP summary with feature names
# shap.summary_plot(shap_values, X_test, feature_names=feature_names, show=False)
# plt.savefig('shap_summary_plot_all_classes.png')
```
